```python
import pickle
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size':20})
```

```python
with(open('./model_history','rb')) as openfile:
    history = pickle.load(openfile)
```

```python
history.keys()
```

```python
training_loss = history['loss']
val_loss = history['val_loss']

epoch_count = range(1, len(training_loss) + 1)
plt.figure(figsize=(14,10))
plt.plot(epoch_count, training_loss, 'r--')
plt.plot(epoch_count, val_loss, 'b-')
plt.legend(['Erro de treinamento', 'Erro de validao'])
plt.xlabel('poca')
plt.ylabel('Erro')
plt.grid()
plt.show()
```

```python
print(f'Erro de treinamento final: {training_loss[-1]}\nErro de validao final: {val_loss[-1]}')
```

```python
training_acc = history['accuracy']
val_acc = history['val_accuracy']

epoch_count = range(1, len(training_acc) + 1)
plt.figure(figsize=(14,10))
plt.plot(epoch_count, training_acc, 'r--')
plt.plot(epoch_count, val_acc, 'b-')
plt.legend(['Acurcia de treinamento', 'Acurcia de validao'])
plt.xlabel('poca')
plt.ylabel('Acurcia')
plt.grid()
plt.show()
```

```python
print(f'Acurcia para o conjunto de treinamento: {training_acc[-1]}\nAcurcia para o conjunto de validao: {val_acc[-1]}')
```
